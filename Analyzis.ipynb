```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
input_data = Path('data/raw/life_expectancy_data.csv')
df = pd.read_csv(input_data)
```

```python
df[df['Country'] == 'Russian Federation'].max().loc['Population']
```

```python
values = []
countries = df['Country'].unique()
for country in countries:
    values.append((df[df['Country'] == country].max().loc['Population']))
bars = plt.bar(countries, values)
plt.xticks(rotation=90)
plt.suptitle('Population of each country')
for bar, value in zip(bars, values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height+5, value, ha='center', va = 'bottom')
plt.show()
```

```python
plt.show()
values = []
statuses = df['Status'].unique()
for status in statuses:
    values.append((df[df['Status'] == status].mean().loc['Life expectancy ']))
bars = plt.bar(statuses, values)
plt.xticks(rotation=90)
plt.suptitle('Status with life expectancy')
for bar, value in zip(bars, values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 5, value, ha='center', va='bottom')
plt.show()
```

# Итоги анализа

По итогу анализа было выявлено, что на показатель "Life Expectancy " влияет статус страны (Developing/Developed). В развивающихся странах показатель ниже. Income composition of resources и Schooling заметно коррелируются с ожидаемой продолжительностью жизни. Также продолжительности жизни увеличивалась с каждым годом во всех странах. Так как для обучения требуются только численные данные, то потребуется перевести строковые переменные в числовой вид. В связи с этим появляется проблема в изменении столбца с наименованиями стран. Для их кодирования можно использовать one hot encoding или передать каждой стране или группе стран уникальный номер. Так как государство можно однозначно определить по ряду параметров (Status, percentage expenditure, Total expenditure, GDP, Population, Income composition of resources),то было принято решение удалить колонку стран и не проводить кодирование их названий. Параметр Status будет изменен в соответствии (Developing – 0, Developed - 1). Пропуски в данных разумно будет заполнить (не для нашего предсказываемого параметра) значениями. Также можно отразить, что в некоторых местах были пропущены значения, что может положительно сказаться на обучении.
